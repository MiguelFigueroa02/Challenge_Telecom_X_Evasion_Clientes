Para empezar, hago un anÃ¡lisis sobre la columna Churn

```python
datos['Churn'].unique()
```

Para detectar valores positivos, negativos y nulos, se convierte resultados a tipo booleano y los resultados vaciios se convierten a null

```python
datos['Churn'] = datos['Churn'].replace('Yes',True).replace('No',False)
```

```python
columnas_customer = list(customer_normalizado.columns)
columnas_customer
for columna in columnas_customer:
  print(f'En la columna {columna} se encuentran los siguientes valores Ãºnicos: {customer_normalizado[columna].unique()}')
```

```python
customer_normalizado['Dependents'] = customer_normalizado['Dependents'].replace('Yes',True).replace('No',False)
```
