```python
def variables_categoricas(variable):
  df_grouped = datos_normalizados.groupby(['Churn',variable]).size().reset_index(name='Cantidad')
  df_grouped['Churn'] = df_grouped['Churn'].replace(0,'No_Churn').replace(1,'Churn')
  df_grouped[variable] = df_grouped[variable].replace(0,f'No_'+ variable).replace(1,variable)
  # print(df_grouped)
  df_pivot = df_grouped.pivot(index=variable, columns='Churn',values='Cantidad').fillna(0)
  df_pivot['Churn']=round((df_pivot['Churn']/df_pivot['Churn'].sum())*100,2)
  df_pivot['No_Churn']=round((df_pivot['No_Churn']/df_pivot['No_Churn'].sum())*100,2)
  # print(df_pivot)
  valores_unicos = df_grouped[variable].unique().tolist()
  # print(valores_unicos)
  colors={valores_unicos[0]:'#2B303A',valores_unicos[1]: '#307351'}
  # print(colors)


  fig = px.sunburst(df_grouped, path=['Churn',variable], values='Cantidad',color='Churn',labels='Churn',
                  color_discrete_map={'No_Churn': '#1984c5', 'Churn': '#c23728'})
  fig.update_layout(
    title= dict(text=f'Distribuci칩n de Churn en Clientes conforme a columna<br>{variable}',x = 0.1,
          xanchor='left'),
    width=600,height=600, paper_bgcolor= "#e2e2e2"
  )
  fig.update_traces(textinfo='label+value',insidetextorientation='radial', rotation= 40)
  fig.update_traces(marker_line_color='White',marker_line_width=1.5)
  fig.show()

  df_pivot = df_pivot.T
  print('-'*150)

  fig_2 = px.bar(df_pivot, y=df_pivot.index, x=df_pivot.columns, barmode='group',text_auto=True, labels={'Churn': 'Cancelaci칩n', 'value': 'Proporci칩n (%)'})
  for trace in fig_2.data:
    trace_name = trace.name
    if trace_name in colors:
        trace.marker.color = colors[trace_name]
  fig_2.update_layout(
      title= dict(text=f'Distribuci칩n de Churn en Clientes conforme a columna<br>{variable}',x = 0.1,
            xanchor='left'),
      plot_bgcolor="#e2e2e2", xaxis= dict(showline=True,
          showgrid=False),
      yaxis= dict(showline=False,
          showgrid=False, tickfont=dict(size=15)),
      width=1100,height=600, paper_bgcolor= "#e2e2e2"
  )
  fig_2.update_traces(textposition='outside')
  fig_2.show()
```

```python
import numpy as np
import plotly.express as px

def ancho_bin(df, columna):
    q75, q25 = np.percentile(df[columna], [75, 25])
    iqr = q75 - q25
    ancho = 2 * iqr * np.power(len(df[columna]), -1/3)
    return ancho
```

```python
def variables_numericas_churn_2(df, variable):
    test_df_2= df[(df['Churn']==True)][variable].value_counts().fillna(0).sort_index().reset_index(0)
    test_df_2.columns= [variable,'Cantidad_2']
    binwidth = ancho_bin(test_df_2, variable)

    xmin = test_df_2[variable].min()
    xmax = test_df_2[variable].max()

    bins = np.arange(xmin, xmax + binwidth, binwidth)
    labels = [f"{int(i)} - {int(i+binwidth)}" for i in bins[:-1]]

    test_df_2['bin'] = pd.cut(
        test_df_2[variable],
        bins=bins,
        labels=labels,
        include_lowest=True
    )

    # Ahora contamos cu치ntos caen en cada rango
    agrupado = test_df_2['bin'].value_counts(normalize=True).sort_index().reset_index()
    agrupado.columns = ['Rango', 'Frecuencia']
    agrupado['Frecuencia']= (agrupado['Frecuencia']*100).round(2)

    # Gr치fico de barras
    fig = px.bar(
        agrupado,
        x="Rango",
        y="Frecuencia",
        title=f"Barras agrupadas por rangos de {variable}",
        color_discrete_sequence=["#76c893"]
    )
    fig.update_layout(
        title= dict(text=f'Distribuci칩n de Churn en Clientes conforme a columna<br>{variable}',
                xanchor='center', x = 0.5, font_size=20),
        plot_bgcolor="#e2e2e2", xaxis= dict(showline=True,
              showgrid=False),
        yaxis= dict(showline=False,
              showgrid=False, tickfont=dict(size=12), range=[0,agrupado['Frecuencia'].max()]),
        width=1100,height=600, paper_bgcolor= "#e2e2e2", showlegend=False, margin=dict(l=20, r=5, b=20, t=60)
    )
    fig.update_xaxes(title_text=f'Rangos de variable {variable}')
    fig.update_yaxes(title_text='Proporci칩n (%)')

    fig.show()
```

```python
def variables_categoricas_2(df,variable):
  df_grouped = df.groupby(['Churn',variable]).size().reset_index(name='Cantidad')
  df_grouped['Churn'] = df_grouped['Churn'].replace(0,'No_Churn').replace(1,'Churn')
  df_grouped[variable] = df_grouped[variable].replace(0,f'No_'+ variable).replace(1,variable)
  # print(df_grouped)
  df_pivot = df_grouped.pivot(index=variable, columns='Churn',values='Cantidad').fillna(0)
  df_pivot['Churn']=round((df_pivot['Churn']/df_pivot['Churn'].sum())*100,2)
  df_pivot['No_Churn']=round((df_pivot['No_Churn']/df_pivot['No_Churn'].sum())*100,2)
  # print(df_pivot)
  valores_unicos = df_grouped[variable].unique().tolist()
  # print(valores_unicos)
  colors={valores_unicos[0]:'#2B303A',valores_unicos[1]: '#307351'}
  # print(colors)


  fig = px.sunburst(df_grouped, path=['Churn',variable], values='Cantidad',color='Churn',labels='Churn',
                  color_discrete_map={'No_Churn': '#1984c5', 'Churn': '#c23728'})
  fig.update_layout(
    title= dict(text=f'Distribuci칩n de Churn en Clientes conforme a columna<br>{variable}',x = 0.1,
          xanchor='left'),
    width=600,height=600, paper_bgcolor= "#e2e2e2"
  )
  fig.update_traces(textinfo='label+value',insidetextorientation='radial', rotation= 40)
  fig.update_traces(marker_line_color='White',marker_line_width=1.5)
  fig.show()

  df_pivot = df_pivot.T
  print('-'*150)

  fig_2 = px.bar(df_pivot, y=df_pivot.index, x=df_pivot.columns, barmode='group',text_auto=True, labels={'Churn': 'Cancelaci칩n', 'value': 'Proporci칩n (%)'})
  for trace in fig_2.data:
    trace_name = trace.name
    if trace_name in colors:
        trace.marker.color = colors[trace_name]
  fig_2.update_layout(
      title= dict(text=f'Distribuci칩n de Churn en Clientes conforme a columna<br>{variable}',x = 0.1,
            xanchor='left'),
      plot_bgcolor="#e2e2e2", xaxis= dict(showline=True,
          showgrid=False),
      yaxis= dict(showline=False,
          showgrid=False, tickfont=dict(size=15)),
      width=1100,height=600, paper_bgcolor= "#e2e2e2"
  )
  fig_2.update_traces(textposition='outside')
  fig_2.show()
```
