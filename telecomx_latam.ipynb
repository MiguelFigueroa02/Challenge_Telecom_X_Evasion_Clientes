Dado que el status de los clientes con valores vac√≠os a√∫n no est√° definido, se toma como si la decisi√≥n de dejar o no la compa√±√≠a no ha sido tomada por los clientes. Por lo cual, estos mantienen el status de sus servicios vigentes hasta que sea tomada una decisi√≥n.

## Evaluaci√≥n de la informaci√≥ adicional

Verifico el tipo de datos √∫nicos para determinar el tratamiento a realizar


```python
columnas_customer = list(datos_normalizados.columns)
columnas_customer
for columna in columnas_customer:
  mensaje = f'En la columna {columna} se encuentran un total de {datos_normalizados[columna].nunique()} valores √∫nicos.'
  if datos_normalizados[columna].nunique() < 50:
    mensaje = mensaje + f'\n Los valores √∫nicos son {datos_normalizados[columna].unique()}'
  print(mensaje)
  print('-'* 50)
```

```python
columnas_normalizadas = list(datos_normalizados.columns)
columnas_normalizadas
```

```python
for columna in columnas_normalizadas:
  valor = None
  valor = datos_normalizados[columna].isin(['',' ',None,'False']).sum()
  if valor > 0:
    print(f'La columna {columna} tiene {valor} valores vac√≠os ''o iguales a none')
```

Eval√∫o el comportamiento sobre la columna account.Charges.Total para los valores vac√≠os.

```python
datos_normalizados[
    (datos_normalizados['account.Charges.Total'].isin(['',' ', False, 'False'])) |
    (datos_normalizados['account.Charges.Total'].isna())
]
```

Puesto que se encuentra una relaci√≥n directa sobre de estas columnas, las cuales tienen una antig√ºedad de 0 meses de contrato, los valores de la columna account.Charges.Total se convertir√°n a cero dado que a√∫n no ha llegado un gasto para estos usuarios.

```python
datos_normalizados['account.Charges.Total'] = datos_normalizados['account.Charges.Total'].apply(lambda x: 0 if str(x).strip() in ['','False',False] else x)
```

Se convierte a tipo num√©rico

```python
import numpy as np
```

```python
datos_normalizados['account.Charges.Total'] = datos_normalizados['account.Charges.Total'].astype(np.float64)
datos_normalizados['account.Charges.Total'].info()
```

Se hace una modificaci√≥n  para que las columnas con valores √∫nicos 'Yes' y 'No' √≥ 0 y 1 se conviertan a tipo booleano, con valores True y False

```python
columnas_bool = ['customer.SeniorCitizen','customer.Partner','customer.Dependents','phone.PhoneService','account.PaperlessBilling']
```
