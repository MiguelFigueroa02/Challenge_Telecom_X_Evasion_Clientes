### Carga de la informaciÃ³n


```python
import pandas as pd
import requests
```

```python
url = 'https://raw.githubusercontent.com/MiguelFigueroa02/Challenge_Telecom_X_Evasion_Clientes/refs/heads/main/TelecomX_Data.json'
data = requests.get(url)
```

```python
from pandas import json_normalize
datos_normalizados = json_normalize(data)
datos_normalizados.info()
```

```python
datos_normalizados['customerID'].duplicated().sum()
```

Para empezar, hago un anÃ¡lisis sobre la columna Churn

```python
datos_normalizados['Churn'].unique()
```

Exploro valores vacÃ­os

```python
datos_normalizados.query('Churn == ""')
```

Dado que el status de los clientes con valores vacÃ­os aÃºn no estÃ¡ definido, se toma como si la decisiÃ³n de dejar o no la compaÃ±Ã­a no ha sido tomada por los clientes. Por lo cual, estos mantienen el status de sus servicios vigentes hasta que sea tomada una decisiÃ³n.

Para detectar valores positivos, negativos y nulos, se convierte resultados a tipo booleano y los resultados vaciios se convierten a null

```python
datos_normalizados['Churn'] = datos_normalizados['Churn'].replace('Yes',True).replace('No',False)
```

```python
datos_normalizados['Churn'] = datos_normalizados['Churn'].replace('',None)
```

## EvaluaciÃ³n de la informaciÃ³ adicional

Verifico el tipo de datos Ãºnicos para determinar el tratamiento a realizar


```python
columnas_customer = list(datos_normalizados.columns)
columnas_customer
for columna in columnas_customer:
  mensaje = f'En la columna {columna} se encuentran un total de {datos_normalizados[columna].nunique()} valores Ãºnicos.'
  if datos_normalizados[columna].nunique() < 50:
    mensaje = mensaje + f'\n Los valores Ãºnicos son {datos_normalizados[columna].unique()}'
  print(mensaje)
  print('-'* 50)
```

Voy a ver cuÃ¡ntas en las cuÃ¡ntos valores vacÃ­os existe en cada una de las columnas


```python
columnas_normalizadas = list(datos_normalizados.columns)
columnas_normalizadas
```

```python
for columna in columnas_normalizadas:
  valor = None
  valor = datos_normalizados[columna].isin(['',' ',None,'False']).sum()
  if valor > 0:
    print(f'La columna {columna} tiene {valor} valores vacÃ­os ''o iguales a none')
```

EvalÃºo el comportamiento sobre la columna account.Charges.Total para los valores vacÃ­os.

```python
datos_normalizados[
    (datos_normalizados['account.Charges.Total'].isin(['',' ', False, 'False'])) |
    (datos_normalizados['account.Charges.Total'].isna())
]
```

Puesto que se encuentra una relaciÃ³n directa sobre de estas columnas, las cuales tienen una antigÃ¼edad de 0 meses de contrato, los valores de la columna account.Charges.Total se convertirÃ¡n a cero dado que aÃºn no ha llegado un gasto para estos usuarios.

```python
datos_normalizados['account.Charges.Total'] = datos_normalizados['account.Charges.Total'].apply(lambda x: 0 if str(x).strip() in ['','False',False] else x)
```

Se convierte a tipo numÃ©rico

```python
import numpy as np
```

```python
datos_normalizados['account.Charges.Total'] = datos_normalizados['account.Charges.Total'].astype(np.float64)
datos_normalizados['account.Charges.Total'].info()
```

Se hace una modificaciÃ³n  para que las columnas con valores Ãºnicos 'Yes' y 'No' Ã³ 0 y 1 se conviertan a tipo booleano, con valores True y False

```python
columnas_bool = ['customer.SeniorCitizen','customer.Partner','customer.Dependents','phone.PhoneService','account.PaperlessBilling']
```

```python
for columna in columnas_bool:
  datos_normalizados[columna] = datos_normalizados[columna].apply(lambda x: True if x in [1,'Yes'] else x).apply(lambda y: False if y in [0,'No'] else y)
```
